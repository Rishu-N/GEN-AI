````python
system_message = "You are a helpful assistant that responds expressively in Markdown"
user_prompt = "How do I decide if a business problem is suitable for an LLM solution?"

prompts = [
    {"role": "system", "content": system_message},
    {"role": "user", "content": user_prompt}
  ]
````

````python
stream = openai.chat.completions.create(
    model='gpt-4o-mini',
    messages=prompts,
    temperature=0.7,
    stream=True
)

reply = ""
display_handle = display(Markdown(""), display_id=True)
for chunk in stream:
    reply += chunk.choices[0].delta.content or ''
    reply = reply.replace("```","").replace("markdown","")
    update_display(Markdown(reply), display_id=display_handle.display_id)
````

````python
# we are going to have a conversation between deepseekr1 and gpt4.1nano

deepseek_role = "you have to keep arguaing against the startup ideas and find \
                loop holes and make reasons as to why they arent feasible\
                be very defensive and mean as well sometimes\
                reply in MARKDOWN expresively using points, table and chart/graphs"
gpt_role = "You are a brilliant entrepreneur and when questioned \
            about it you have to defend it while being nice\
            reply in MARKDOWN expresively using points, table and chart/graphs"
````

````python
deepseek_prompt = [
    {"role":"system","content":deepseek_role}
]
gpt_prompt = [
    {"role":"system","content":gpt_role},
    {"role":"user","content":"I want some startup ideas"}
]
````

````python
for i in range(3):
    # GPT CALL
    gpt_stream = openai.chat.completions.create(
        model = chatgpt_models[5],
        messages = gpt_prompt,
        stream = True
    )
    print(f"\n----------------------------\n[{i}] GPT RESPONCE\n------------------------\n")
    reply = ""
    display_handle = display(Markdown(""), display_id=True)
    for chunk in gpt_stream:
        reply += chunk.choices[0].delta.content or ''
        reply = reply.replace("```","").replace("markdown","")
        update_display(Markdown(reply), display_id=display_handle.display_id)

    gpt_responce = reply
    
    deepseek_prompt.append(
        {"role":"user","content":gpt_responce}
    )
    gpt_prompt.append(
        {"role":"assistant","content":gpt_responce}
    )

    #DEEPSEEK CALL
    deepseek_reply = ollama.chat(
        model = "deepseek-r1:1.5b",
        messages = deepseek_prompt,
    )
    deepseek_responce = deepseek_reply['message']['content']
    print(f"\n----------------------------\n[{i}] DEEPSEEK RESPONCE\n------------------------\n")
    display(Markdown(deepseek_responce))
    
    gpt_prompt.append(
        {"role":"user","content":deepseek_responce}
    )
    deepseek_prompt.append(
        {"role":"assistant","content":deepseek_responce}
    )
    
````

````python
analysis_prompt = gpt_prompt
analysis_prompt.append({"role":"user","content":"analyze the conversation above and derive insights"})

for part in ollama.chat(model = "qwen3:4b", messages=analysis_prompt, stream=True):
    print(part['message']['content'], end='', flush=True)
````
